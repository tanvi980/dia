```python
df1=df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
df1.values[0,0:2]
df1
```

```python

sns.barplot(data=df1,y=df1.values[0:2,0],x=df1.index)
plt.ylabel('math score')
```

- Cource='math score'
   - 69.68% students have completed their math cource
   - 64.73% students have not completed their math score

```python
sns.barplot(data=df1,y=df1.values[0:2,1],x=df1.index)
plt.ylabel('reading score')
```

- Cource='reading score'
   - 74.08% students have completed their reading cource
   - 66.43% students have not completed their reading score

```python
sns.barplot(data=df1,y=df1.values[0:2,2],x=df1.index)
plt.ylabel('writing score')
```

- Cource='writing score'
   - 74.67% students have completed their writing cource
   - 64.24% students have not completed their writing score

```python
plt.figure(figsize=(25,25))
df1=df.groupby('parental level of education')[['math score','reading score','writing score']].mean()
print(df1)

df1.plot(kind='bar')
```
