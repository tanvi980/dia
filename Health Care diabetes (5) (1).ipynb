```python
df['Glucose']=df['Glucose'].replace(0,'NAN')
df['BloodPressure']=df['BloodPressure'].replace(0,'NAN')
df['SkinThickness']=df['SkinThickness'].replace(0,'NAN')
df['Insulin']=df['Insulin'].replace(0,'NAN')
df['BMI']=df['BMI'].replace(0,'NAN')
df['DiabetesPedigreeFunction']=df['DiabetesPedigreeFunction'].replace(0,'NAN')
```

```python
df[{'Glucose','BloodPressure','SkinThickness','Insulin','BMI','Pregnancies'}].astype(float)
```

 By using .astype() function we convert all the values of glucose,skin thickness ,BMI, blood pressure ,pregnancies and insulin into float

```python
df['Glucose']=df['Glucose'].astype(float)
df['Glucose'].fillna(df['Glucose'].mean(),inplace=True)
df['Insulin']=df['Insulin'].astype(float)
df['Insulin'].fillna(df['Insulin'].mean(),inplace=True)
df['SkinThickness']=df['SkinThickness'].astype(float)
df['SkinThickness'].fillna(df['SkinThickness'].mean(),inplace=True)
df['BloodPressure']=df['BloodPressure'].astype(float)
df['BloodPressure'].fillna(df['BloodPressure'].mean(),inplace=True)
df['BMI']=df['BMI'].astype(float)
df['BMI'].fillna(df['BMI'].mean(),inplace=True)
df['Pregnancies']=df['Pregnancies'].astype(float)
df['Pregnancies'].fillna(df['Pregnancies'].mean(),inplace=True)
```

```python
x=df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=11)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print(metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
print(metrics.accuracy_score(y_test,y_pred))
```
